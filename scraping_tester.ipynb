```python
soup2 = BeautifulSoup(urlopen(url), 'lxml')
```

```python
print(soup2.title.text)
print()
print(soup2.abstract.text)
```

```python
description = soup2.find_all('div', attrs={'class':"description-line"})

for d in description:
    d = d.get_text()
    print (d)
```

```python
def claim_finder(url):
    #function for extracting all the text from the Claim Section
    #for some reason, it is printing the first claim twice
    claim = soup2.find_all('div', attrs={'class':"claim"})
    for c in claim:
        c = c.get_text()
        print(c)
```

```python
claim_finder(url)
```
